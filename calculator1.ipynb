```python
print ('welcome to my simple calculator 0_0  ')

num_one = int(input("Enter first number: "))

op = input("Enter operator: ")

num_two = int(input("Enter second number: "))

if op == "+":
    print(num_one + num_two)
elif op == "-":
    print(num_one - num_two)
elif op == "*" or op == "x":
    print(num_one * num_two)
elif op == "/":
    print(num_one / num_two)
```
